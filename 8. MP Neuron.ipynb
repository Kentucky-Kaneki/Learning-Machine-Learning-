```python
def mp_neuron(inputs, weights, threshold):
    # Calculate weighted sum
    weighted_sum = sum(i * w for i, w in zip(inputs, weights))
    # Output 1 if weighted sum >= threshold else 0
    return 1 if weighted_sum >= threshold else 0

def get_inputs(n):
    inputs = []
    for i in range(n):
        while True:
            val = input(f"Enter input {i+1} (0 or 1): ")
            if val in ['0', '1']:
                inputs.append(int(val))
                break
            else:
                print("Invalid input, please enter 0 or 1.")
    return inputs

def and_gate():
    n = int(input("Enter the number of inputs for AND gate: "))
    inputs = get_inputs(n)
    weights = [1] * n
    threshold = n  # For AND, threshold = number of inputs
    output = mp_neuron(inputs, weights, threshold)
    print(f"AND gate output: {output}")

def or_gate():
    n = int(input("Enter the number of inputs for OR gate: "))
    inputs = get_inputs(n)
    weights = [1] * n
    threshold = 1  # For OR, threshold = 1
    output = mp_neuron(inputs, weights, threshold)
    print(f"OR gate output: {output}")

def main():
    while True:
        choice = input("Choose gate to simulate (AND/OR) or Q to quit: ").strip().upper()
        if choice == 'AND':
            and_gate()
        elif choice == 'OR':
            or_gate()
        elif choice == 'Q':
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please enter AND, OR, or Q.")

if __name__ == "__main__":
    main()
```
